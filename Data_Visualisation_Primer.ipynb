# Now let's see line-plot:--->

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

# Now for line plot we have to use the function named as plot()...so let's have some knowledge about the plot()

```python
plt.plot?
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.figure(figsize=(15,8))
plt.plot(x,y1,color='red',label='Apple',marker='o',linestyle='dashed')
plt.plot(x,y2,color='green',label='Kiwi',marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price of Fruits over time")
plt.legend()
plt.show()
```

# plt.show()--> tells us that we are done with all doing on the specific graph and we need to display it! If you use 2- plt.show() then 2-graph page will be visible.

# plt.plot(only-single-argument)--> If you don't give x-axis the value then the given value will be automatically assigned to the y-axis while index of the array becomes x-axis... 

# Now let's see Scatter-plot:--->

```python
print(x)
print(y1)
print(y2)
```

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-dark')
plt.figure(figsize=(10,8))
plt.scatter(x,y1,color='red',label='Apple',marker='o')
plt.scatter(x,y2,color='green',label='kiwi',marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price of Fruits over time")
plt.legend()
plt.show()
```

# Now let's see Bar-plot:--->

```python
plt.style.use('classic')
x_coor = np.array([0,1,2])*2
plt.bar(x_coor-0.25,[10,20,15],width=0.5,label='Current Year',tick_label=['Gold','Platinum','Rose'])
plt.bar(x_coor+0.25,[20,10,12],width=0.5,label='Next Year')
plt.ylim(0,40)
plt.xlim(-2,6)
plt.legend()
plt.show()
```

# Now let's see Pie-Chart:--->

```python
plt.style.use('classic')
subjects = "Maths","Chemistry","Physics","English"
weights = [20,10,15,5]
plt.pie(weights,labels=subjects,autopct='%1.1f%%',explode=(0,0,0.25,0),shadow=True)
plt.show()
```
